```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
# Camp century
cc_1977 = pd.read_csv("data/campcentury/del1977.txt", sep="	",skiprows=74, names=["year", "d18o_cc1977"])
cc_1977 = cc_1977.set_index('year')
cc_latlon = (77.167,-61.133)

# longer series past 1700s
cc_1967 = pd.read_csv("data/campcentury/cc-1ynew.txt", sep="	",skiprows=75, names=["year", "d18o_cc1967"])
cc_1967 = cc_1967.set_index('year')

# Dye 2
dye2 = pd.read_csv("data/dye2/dye2.csv",names=["year", "d18o_dye2"])[:-1]
dye2 = dye2.set_index('year')

# Dye 3
dye3 = pd.read_csv("data/dye3/dye3-1yr.txt", sep="	",skiprows=17, names=["year", "d18o_dye3"])[:630]
dye3 = dye3.set_index('year')

# Summit
summit = pd.read_csv("data/summit/sum74-1y.txt", sep="	",skiprows=79, names=["year", "d18o_summit"])
summit = summit.set_index('year')
```

```python
all_cores = pd.concat([cc_1977,cc_1967,dye2,dye3,summit],axis=1).sort_index()
all_cores.plot()
plt.xlim([1800,2000])
```
